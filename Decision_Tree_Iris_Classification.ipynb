# ðŸŒ³ Decision Tree Classification using scikit-learn

This notebook demonstrates how to build and visualize a Decision Tree Classifier using the famous **Iris dataset**.

We will:
- Load the dataset
- Train a decision tree model
- Visualize the tree
- Evaluate the model


```python
# Importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
# Load the iris dataset
iris = load_iris()

# Create a DataFrame from the dataset
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Add the target column to the DataFrame
df['target'] = iris.target

# Show the first 5 rows
df.head()
```

```python
# Check dataset structure
df.info()
```

```python
# Define features and labels
X = df[iris.feature_names]  # input columns
y = df['target']            # target (flower type)

# Split the dataset into 70% training and 30% testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Create the decision tree classifier
model = DecisionTreeClassifier()

# Train it on training data
model.fit(X_train, y_train)
```

```python
# Visualize the decision tree
plt.figure(figsize=(20, 10))
plot_tree(model, filled=True, feature_names=iris.feature_names, class_names=iris.target_names)
plt.title("Decision Tree for Iris Dataset")
plt.show()
```

```python
# Predict using the testing data
y_pred = model.predict(X_test)

# Print Accuracy
print("Accuracy:", accuracy_score(y_test, y_pred))

# Confusion Matrix
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))

# Classification Report
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

## ðŸ“Š Analysis

- The dataset contains 3 classes of flowers: Setosa, Versicolor, and Virginica.
- The decision tree model was trained on 70% of the data and tested on the remaining 30%.
- The tree learned how to separate the flowers based on their petal and sepal measurements.
- The model achieved high accuracy, showing that the decision tree is well-suited for this dataset.
- Visualization helps us understand how the model is making decisions.
